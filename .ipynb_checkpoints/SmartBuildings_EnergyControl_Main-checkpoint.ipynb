# RL and environment classes:
Here we briefly explain how different RL and environment classes work and what inputs do they require.

## Environment classes:
We have two environment classes called Simple1Room and Simple1Room_gym. The latter is the gym-compatible version of the former that we later use to implement the PPO algorithm from Stabe Baselines3.
### 1. Simple1Room:
This is the environment class for a simple linear model of a one-zone buidling/room. The linear model is as follows:

$\frac{dT}{dt} = \alpha (T-T_o) + h_s\dot{Q}_h $,

where $\alpha=K/C$, and $\dot{Q}_h=\dot{q}_h/C$. Here, $C=2000 \, kJK^{-1}$ is the building's heat capacity, $K=325 WK^{-1}$ is building's thermal conductance, and $\dot{q}_h=13 \, kW$ is the heater's power. Also, $h_s(t) \in \{0,1\}$ is the heater status, and $T_o=-10 \, ^\circ C$ is the outdoor temperature. The numerical values mentioned here are the default values and they could be overwritten by the class's input.

Rewards are calculated in the environment class during simulation of each transition. It consists of two time-continuous and one discrete components:

\begin{equation}
r_{k+1}=r_{sw}+\int_{t_k^+}^{t_{k+1}^+}r_e h_s(t)+r_c f(T,T_d) \: dt,
\end{equation}

where,

\begin{equation}
    f(T,T_d)=
    \begin{cases}
      0, & \text{if}\ |T-T_d|<=1 \\
      (|T-T_d|-1)^2, & \text{otherwise}
    \end{cases}
  \end{equation}
  
## RL classes:
We have four main classes for RL, namely, Stochastic_AC_PG and COPDAC_Q, COPDAC_Q_Ton, and PPO. The first 3 are developed and coded by me, while the last is adopted from the Stable Baselines3 implementation. 

### 1. Stochastic_AC_PG
In this class, the action (temperature threshold) is sampled from a Gaussian distribution. Mean and standard deviation of this distribution is parameterized which the algorithm tries to learn.

### 2. COPDAC_Q
This class is a compatible off-policy deterministic actor-critic RL with simple Q-learning critic. Here we define the deterministic policy as $a=\mu_{\theta}(s)=[1-h_s, h_s][T^{\mathrm{th}}_{\mathrm{ON}}, T^{\mathrm{th}}_{\mathrm{OFF}}]^\top$. We further simplify it by assuming switching thresholds are constants and do not depend on the outdoor temperature; $T^{\mathrm{th}}_{\mathrm{ON}}\doteq \theta_{\mathrm{ON}}$ and $T^{\mathrm{th}}_{\mathrm{OFF}}\doteq \theta_{\mathrm{OFF}}$. This is a reasonable assumption because we know the outdoor temperature is fixed and the optimal thresholds are indeed constant. We can write the policy as:

 \begin{equation}
     \mu_{\theta}(s) \doteq \theta^\top \phi(s),
     \label{Eq:policy}
 \end{equation}
 
where, $\theta=[\theta_{\mathrm{ON}}, \theta_{\mathrm{OFF}}]^\top$ is the policy parameter vector, and $\phi(s)=[1-h_s, h_s]^\top$ is the state feature vector. We approximate the state-value function as $V_v(s)=[v_1, v_2]^\top \phi(s)$ and the action-value function by a compatible function approximator as $Q^w(s,a)=(a-\mu_{\theta}(s))^\top{\nabla_{\theta}\mu_{\theta}(s)}^\top w + V^v(s)$ with $w=[w_1, w_2]^\top$.

By default the [MDP] transitions have variable time interval (transitions happen when hitting switching manifolds by default); however, COPDAC_Q can handle fixed transition intervals as well, where tarnsitions are defined at fixed time intervals.

### 3. COPDAC_Q_Ton

This class is the simplified version of COPDAC_Q where the switching-off threshold is fixed and the agent only learns the switching-on threshold temperature.

### 4. PPO

This is the off-the-shelf Proximal Policy Optimization algorithm imported from Stable Baselines 3.
https://stable-baselines3.readthedocs.io/en/master/modules/ppo.html
***

## A. Groundtruth optimal policy
Because there is no stochasticity in the simple version of the problem and that the outdoor temperature is fixed the optimal policy within the the threshold-family policies should indeed be a deterministic policy with constant temperature switching thresholds. 

Therefore, in order to find the optimal values of these thresholds we simulate the system for a range of deterministic temperature thresholds for a long time and calculate their corresponding average reward calculated as the total accumulated reward divided by the total time.

```python
# below we define the function that takes an env object and two arrays for
# switching ON and OFF temperature thresholds as well as two options to whether
# plot a high-quality figure and to whether save the HQ figures or not


def groundtruth_simulation(env, TON_range, TOFF_range, HQ_fig, save_fig):

    Rave_mat = np.zeros((TOFF_range.size, TON_range.size))

    x_ind = -1
    for TON in TON_range:
        x_ind = x_ind + 1
        y_ind = -1
        for TOFF in TOFF_range:
            y_ind = y_ind + 1
            Rave_mat[y_ind, x_ind] = env._RaveCalc(theta_mu=np.array([TON, TOFF]),
                                                   theta_sigma=1,
                                                   T_lowerbound=env.Td - 5,
                                                   T_upperbound=env.Td + 5,
                                                   deterministic=True,
                                                   max_iter=500,
                                                   max_hour=5*24.0,
                                                   plots=False,
                                                   PerTimeStep=True)
    plt.close("all")

    ind_max = np.unravel_index(np.argmax(Rave_mat, axis=None), Rave_mat.shape)
    print("The maximum R_ave = {0:.3f}, occurs for T_ON = {1:.2f}, and T_OFF = {2:.2f}"\
          .format(Rave_mat[ind_max[0], ind_max[1]], TON_range[ind_max[1]], TOFF_range[ind_max[0]]))

    if HQ_fig:
        mpl.rcParams['text.usetex'] = True
        fig = plt.figure(figsize=(10, 8))
        ax = fig.gca(projection='3d')
        X_TON, Y_TOFF = np.meshgrid(TON_range, TOFF_range)
        surf = ax.plot_surface(X_TON,
                               Y_TOFF,
                               Rave_mat,
                               cmap=cm.coolwarm,
                               antialiased=False,
                               edgecolor='black',
                               linewidth=0.01)
        ax.contour(X_TON,
                   Y_TOFF,
                   Rave_mat, [-5.5, -5, -4.5, -4, -3.8, -3.75, -3.71],
                   cmap=cm.coolwarm,
                   linestyles="solid",
                   offset=-7.0)
        
        AxesLabelFont = FontProperties(family='Arial',size=20)
        TickLabelFont = FontProperties(family='Arial',size=18)

        
        ax.set_xlabel(r'$T_{\mathrm{ON}}^{\mathrm{th}}(^{\circ}C)$ ',
                      fontproperties=AxesLabelFont,
                      labelpad=18)
        ax.set_ylabel(r'$T_{\mathrm{OFF}}^{\mathrm{th}}(^{\circ}C)$ ',
                      fontproperties=AxesLabelFont,
                      labelpad=18)
        ax.set_zlabel(r'$J$ (unit/hour)',
                      fontproperties=AxesLabelFont,
                      labelpad=12)

        ax.set_xticks([10, 11, 12, 13, 14])
        ax.set_yticks([16, 17, 18, 19, 20])
        ax.set_zlim(-7.0, -4.0)

        # make the panes transparent
        ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
        ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
        ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))

        ax.yaxis._axinfo['label']['space_factor'] = 5.0
        # change fontsize
        a = plt.gca()
        a.set_xticklabels(a.get_xticks(),fontproperties=TickLabelFont)
        a.set_yticklabels(a.get_yticks(),fontproperties=TickLabelFont)
        a.set_zticklabels(a.get_zticks(),fontproperties=TickLabelFont)


        # set the view angle
        ax.view_init(20, None)
        plt.show()

        # saving png and pdf the fig
        if save_fig:
            fig.savefig('groundtruth.png', dpi=200, bbox_inches='tight')
            fig.savefig('groundtruth.pdf', bbox_inches='tight')
        
    return Rave_mat


env_gt = Simple1Room(Td=15.0)
TON_range = np.arange(env_gt.Td - 5, env_gt.Td - 1, 0.1)
TOFF_range = np.arange(env_gt.Td + 1, env_gt.Td + 5, 0.1)

Rave_mat = groundtruth_simulation(env_gt, TON_range, TOFF_range, HQ_fig=True, save_fig=True)
```

## B. Implementing Stochastic_AC_PG on Simple1Room with variable transition periods 

So here we implement an stochastic actor-critic policy gradient algorithm with eligibility traces defined by the *Stochastic_AC_PG* class.

The environment object is defined by the *Simple1Room* class.

In order to compare directly with the groundtruth optimal policy we use the saved env in section A. (we use the same env for all the subsequent RL algorithms unless otherwise specified). We just need to reset the env everytime we use it with a new RL object.


```python
# below we define a function that trains an stochastic RL algorithm on the
# environment env for (max_train_days) days or (max_train_steps) steps whichever
# occurs first.
# if training_plots = True it shows figures related to the training simulation
# if trained_plots = True it shows figures related to a simulation with the
# trained policy

def Stochastic_PG_Simple1Room(env,
                              RL,
                              max_train_days,
                              max_train_steps,
                              training_plots=True,
                              trained_plots=False,
                              HQ_fig=False,
                              save_fig=False):

    env.reset()

    # training the RL object on the env object
    Rave_stepseries, theta_mu_stepseries, theta_sigma_stepseries,\
    w_valuefunc_stepseries, t_stepseries, t_timeseries, T_stepseries, T_timeseries,\
    hs_stepseries, G_stepseries, G_timeseries\
    = train(env, RL, max_simulation_days = max_train_days, max_steps = max_train_steps)

    t_stepseries[:] = [x / 3600 for x in t_stepseries]
    t_timeseries[:] = [x / 3600 for x in t_timeseries]
    if RL.PerTimeStep:
        Rave_stepseries[:] = [x * 3600 for x in Rave_stepseries]

    theta_ON_stepseries = [item[0] for item in theta_mu_stepseries]
    theta_OFF_stepseries = [item[1] for item in theta_mu_stepseries]
    w_OFF_stepseries = [item[0] for item in w_valuefunc_stepseries]
    w_ON_stepseries = [item[1] for item in w_valuefunc_stepseries]
    sigma_stepseries = [np.exp(item) for item in theta_sigma_stepseries]

    # taking parameters of the trained policy
    theta_mu_final = theta_mu_stepseries[-1]
    theta_sigma_final = theta_sigma_stepseries[-1]

    # values for plotting probability density functions
    Ton_mean_init = theta_ON_stepseries[0]
    Ton_mean_final = theta_ON_stepseries[-1]
    Toff_mean_init = theta_OFF_stepseries[0]
    Toff_mean_final = theta_OFF_stepseries[-1]
    sigma_init = sigma_stepseries[0]
    sigma_final = sigma_stepseries[-1]
    
    plt.close("all")
    if training_plots:
        mpl.rcParams['text.usetex'] = False
        # Plotting some figures of the training process
        fig, ax1 = plt.subplots(figsize=(16, 6))
        color = 'tab:blue'
        ax1.set_xlabel('time (hour)')
        ax1.set_ylabel('Temperature', color=color)
        ax1.plot(t_timeseries, T_timeseries, color=color)
        ax1.plot(t_stepseries, T_stepseries, '--o', color=color, markersize=6)
        ax1.tick_params(axis='y', labelcolor=color)

        ax2 = ax1.twinx(
        )  # instantiate a second axes that shares the same x-axis
        color = 'tab:red'
        ax2.set_ylabel('switching signal',
                       color=color)  # we already handled the x-label with ax1
        ax2.step(t_stepseries, hs_stepseries, where='post', color=color)
        ax2.tick_params(axis='y', labelcolor=color)
        fig.tight_layout()  # otherwise the right y-label is slightly clipped

        fig, ax1 = plt.subplots(figsize=(8, 6))
        color = 'tab:blue'
        ax1.set_xlabel('time (hour)')
        ax1.set_ylabel('Theta')
        ax1.plot(t_stepseries,
                 theta_ON_stepseries,
                 color=color,
                 label='Theta_ON')
        ax1.plot(t_stepseries,
                 theta_OFF_stepseries,
                 color='tab:red',
                 label='Theta_OFF')
        ax1.tick_params(axis='y')
        ax1.legend()
        fig.tight_layout()  # otherwise the right y-label is slightly clipped

        fig, ax1 = plt.subplots(figsize=(8, 6))
        color = 'tab:blue'
        ax1.set_xlabel('time (hour)')
        ax1.set_ylabel('Standard deviation in Theta')
        ax1.plot(t_stepseries, sigma_stepseries, color=color)
        ax1.tick_params(axis='y')
        fig.tight_layout()

        x = np.linspace(8, 22, 500)
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.set_xlabel('Temperature (deg C)')
        ax.set_ylabel('Probability density')
        ax.plot(x,
                stats.norm.pdf(x, Ton_mean_init, sigma_init),
                color='C0',
                linestyle='-.',
                label='T_ON(initial)')
        ax.plot(x,
                stats.norm.pdf(x, Ton_mean_final, sigma_final),
                color='C0',
                label='T_ON(final)')
        ax.plot(x,
                stats.norm.pdf(x, Toff_mean_init, sigma_init),
                color='C3',
                linestyle='-.',
                label='T_OFF(initial)')
        ax.plot(x,
                stats.norm.pdf(x, Toff_mean_final, sigma_final),
                color='C3',
                label='T_OFF(final)')
        # ax.tick_params(axis='y')
        ax.legend()
        fig.tight_layout()

        fig, ax1 = plt.subplots(figsize=(8, 6))
        color = 'tab:blue'
        ax1.set_xlabel('time (hour)')
        ax1.set_ylabel('w')
        ax1.plot(t_stepseries, w_OFF_stepseries, color=color, label='w_OFF')
        ax1.plot(t_stepseries, w_ON_stepseries, color='tab:red', label='w_ON')
        ax1.tick_params(axis='y')
        ax1.legend()
        fig.tight_layout()  # otherwise the right y-label is slightly clipped

        fig, ax1 = plt.subplots(figsize=(8, 6))
        color = 'tab:blue'
        ax1.set_xlabel('time (hour)')
        if RL.PerTimeStep:
            ax1.set_ylabel('R_ave (per hour)')
        else:
            ax1.set_ylabel('R_ave (per step)')
        ax1.plot(t_stepseries, Rave_stepseries, color=color)
        ax1.tick_params(axis='y')
        fig.tight_layout()
        plt.show()

    if HQ_fig:
        
        plt.close("all")
        mpl.rcParams['text.usetex'] = True
        AxesLabelFont = FontProperties(family='Arial', size=22)
        TickLabelFont = FontProperties(family='Arial', size=22)

        x = np.linspace(8, 22, 800)
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.set_xlabel(r'Temperature $(^{\circ}C)$',
                      fontproperties=AxesLabelFont)
        ax.set_ylabel('Probability density', fontproperties=AxesLabelFont)
        ax.plot(x,
                stats.norm.pdf(x, Ton_mean_init, sigma_init),
                color='C0',
                linestyle='-.',
                linewidth=3.0,
                label=r'$T^{\mathrm{th}}_{\mathrm{ON}}$ (initial)')
        ax.plot(x,
                stats.norm.pdf(x, Ton_mean_final, sigma_final),
                color='C0',
                linewidth=3.0,
                label=r'$T^{\mathrm{th}}_{\mathrm{ON}}$ (learnt)')
        ax.plot(x,
                stats.norm.pdf(x, Toff_mean_init, sigma_init),
                color='C3',
                linestyle='-.',
                linewidth=3.0,
                label=r'$T^{\mathrm{th}}_{\mathrm{OFF}}$ (initial)')
        ax.plot(x,
                stats.norm.pdf(x, Toff_mean_final, sigma_final),
                color='C3',
                linewidth=3.0,
                label=r'$T^{\mathrm{th}}_{\mathrm{OFF}}$ (learnt)')
        plt.legend(fontsize=21)

        a = plt.gca()
        a.set_xticklabels(a.get_xticks(), fontproperties=TickLabelFont)
        a.set_yticklabels(a.get_yticks(), fontproperties=TickLabelFont)
        fig.tight_layout()

        if save_fig:
            fig.savefig('probdist.png', dpi=200, bbox_inches='tight')
            fig.savefig('probdist.pdf', bbox_inches='tight')

        fig, ax1 = plt.subplots(figsize=(10, 6))
        ax1.set_xlabel('time (hour)', fontproperties=AxesLabelFont)
        ax1.set_ylabel(r'Temperature $(^{\circ}C)$',
                       fontproperties=AxesLabelFont)
        plt1 = ax1.plot(t_stepseries,
                        theta_ON_stepseries,
                        color="C0",
                        linewidth=3.0,
                        label=r'$\theta_{\mathrm{ON}}^{m}$')
        plt2 = ax1.plot(t_stepseries,
                        theta_OFF_stepseries,
                        color="C3",
                        linewidth=3.0,
                        label=r'$\theta_{\mathrm{OFF}}^{m}$')
        
        a = plt.gca()
        a.set_xticklabels(a.get_xticks(), fontproperties=TickLabelFont)
        a.set_yticklabels(a.get_yticks(), fontproperties=TickLabelFont)

        ax2 = ax1.twinx(
        )  # instantiate a second axes that shares the same x-axis
        ax2.set_ylabel(r'Standard deviation $(^{\circ}C)$',
                       color='C2',
                       fontproperties=AxesLabelFont
                       )  # we already handled the x-label with ax1
        plt3 = ax2.plot(t_stepseries,
                        sigma_stepseries,
                        color='C2',
                        linewidth=3.0,
                        label=r'$\mathrm{exp}({\theta_{\sigma}})$')
        a = plt.gca()
        a.set_yticklabels(a.get_yticks(), fontproperties=TickLabelFont)
        ax2.tick_params(axis='y', labelcolor='C2')

        # added these three plots together for the unified legend
        pltall = plt1 + plt2 + plt3
        labs = [l.get_label() for l in pltall]
        ax1.legend(pltall, labs, fontsize=21)

        if save_fig:
            fig.savefig('theta_sigma_timehist.png',
                        dpi=200,
                        bbox_inches='tight')
            fig.savefig('theta_sigma_timehist.pdf', bbox_inches='tight')

        fig, ax = plt.subplots(figsize=(10, 6))
        ax.set_ylabel(r'Average reward rate $\bar{r}$ (unit/hour)',
                      fontproperties=AxesLabelFont)
        ax.set_xlabel('time (hour)', fontproperties=AxesLabelFont)
        ax.plot(t_stepseries, Rave_stepseries, color='C0', linewidth=3.0)
        a = plt.gca()
        a.set_xticklabels(a.get_xticks(), fontproperties=TickLabelFont)
        a.set_yticklabels(a.get_yticks(), fontproperties=TickLabelFont)
        fig.tight_layout()
        plt.show()

        if save_fig:
            fig.savefig('AveReward_timehist.png', dpi=200, bbox_inches='tight')
            fig.savefig('AveReward_timehist.pdf', bbox_inches='tight')


    Rave_final_policy=env_gt._RaveCalc(theta_mu_final, theta_sigma_final, T_lowerbound = RL.Tlow,\
              T_upperbound = RL.Tup, max_iter = 100, deterministic = False,\
              PerTimeStep = RL.PerTimeStep, plots = trained_plots)

    print("on-policy Rave (at the end of training simulation) = {}".format(
        Rave_stepseries[-1]))
    print("optimal theta_ON found = {}".format(theta_mu_final[0]))
    print("optimal theta_OFF found = {}".format(theta_mu_final[1]))
    print("optimal (final) standard deviation (of theta) found = {}".format(
        sigma_stepseries[-1]))
    print("Rave using the trained stochastic policy = {}".format(
        Rave_final_policy))
    

    return t_stepseries, Rave_stepseries, theta_ON_stepseries, theta_OFF_stepseries, sigma_stepseries


env_gt = Simple1Room(Td=15.0)
RL = Stochastic_AC_PG(Td=env_gt.Td,
                      learning_rate_policy_mu=0.05,
                      learning_rate_policy_sigma=0.05,
                      learning_rate_valuefunc=0.05,
                      learning_rate_avereward=0.05,
                      elig_decay_policy=0.2,
                      elig_decay_valuefunc=0.2,
                      T_upperbound=22.0,
                      T_lowerbound=8.0)

t_stepseries, Rave_stepseries, theta_ON_stepseries, theta_OFF_stepseries, sigma_stepseries = Stochastic_PG_Simple1Room(
    env_gt,
    RL,
    max_train_days=10,
    max_train_steps=500,
    training_plots=False,
    trained_plots=False,
    HQ_fig=True,
    save_fig=True)
```

## C. Implementing COPDAC_Q on Simple1Room with variable or fixed transition periods
Here we implement a deterministic actor-critic policy defined by the *COPDAC_Q* class with variable transition intervals.

The environment object is defined by the *Simple1Room* class.

In order to compare directly with the groundtruth optimal policy we use the saved env in section A. We just need to reset the env everytime we use it with a new RL object.


```python
# below we define a function that trains a deterministic RL agorithm (COPDAC_Q)
# with variable tarnsition periods on the environment env for (max_train_days) 
# days or (max_train_steps) steps whichever occurs first.
# if training_plots = True it shows figures related to the training simulation
# if trained_plots = True it shows figures related to a simulation with the 
# trained policy
def COPDAC_Simple1Room(env, RL, max_train_days=5, max_train_steps=3000,
                       fixed_TransTime=False, dt_TransStep=5*60,
                       training_plots=True, trained_plots=False,
                       trained_max_hour=5*24.0, HQ_fig=False, save_fig=False,
                       return_arg=False):
    
    envn=copy.deepcopy(env)
    if fixed_TransTime:
        envn.fixed_TransTime=fixed_TransTime
        envn.dt_TransStep=dt_TransStep
        print("\n fixed transition time:")
    else:
        print("\n variable transition time:")
    
    print("---------------------")
    
    envn.reset()

    # training
    Rave_stepseries, theta_stepseries, w_stepseries, v_stepseries, t_stepseries,\
    t_timeseries,T_stepseries, T_timeseries, hs_stepseries, G_stepseries, G_timeseries\
    = train_deterministic(envn, RL, max_simulation_days = max_train_days,
                          max_steps = max_train_steps)

    # converting time from sec to hour
    t_stepseries[:] = [x / 3600 for x in t_stepseries]
    t_timeseries[:] = [x / 3600 for x in t_timeseries]
    if RL.PerTimeStep:
        Rave_stepseries[:] = [x * 3600
                              for x in Rave_stepseries]  # [Rave]=unit/hr

    theta_ON_stepseries = [item[0] for item in theta_stepseries]
    theta_OFF_stepseries = [item[1] for item in theta_stepseries]
    w1_stepseries = [item[0] for item in w_stepseries]
    w2_stepseries = [item[1] for item in w_stepseries]
    v1_stepseries = [item[0] for item in v_stepseries]
    v2_stepseries = [item[1] for item in v_stepseries]

    # final policy parameters
    theta_final = theta_stepseries[-1]
    
    plt.close("all")
    if training_plots:
        
        # plots
        mpl.rcParams['text.usetex'] = False

        fig, ax1 = plt.subplots(figsize=(16, 6))
        color = 'tab:blue'
        ax1.set_xlabel('time (hour)')
        ax1.set_ylabel('Temperature', color=color)
        ax1.plot(t_timeseries, T_timeseries, color=color)
        ax1.plot(t_stepseries, T_stepseries, '--o', color=color, markersize=6)
        ax1.tick_params(axis='y', labelcolor=color)

        ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
        color = 'tab:red'
        ax2.set_ylabel('switching signal',
                       color=color)  # we already handled the x-label with ax1
        ax2.step(t_stepseries, hs_stepseries, where='post', color=color)
        ax2.tick_params(axis='y', labelcolor=color)
        
        fig.tight_layout()  # otherwise the right y-label is slightly clipped

        fig, ax1 = plt.subplots(figsize=(8, 6))
        color = 'tab:blue'
        ax1.set_xlabel('time (hour)')
        if RL.PerTimeStep:
            ax1.set_ylabel('R_ave (per hour)')
        else:
            ax1.set_ylabel('R_ave (per step)')
        ax1.plot(t_stepseries, Rave_stepseries, color=color)
        ax1.tick_params(axis='y')
        fig.tight_layout()
        plt.show()

    Rave_with_theta_final = envn._RaveCalc(theta_final, 1, T_lowerbound = RL.Tlow,
                                          T_upperbound = RL.Tup, max_hour = trained_max_hour,
                                          deterministic = True, 
                                          PerTimeStep = RL.PerTimeStep,
                                          plots = trained_plots)
    
    print("on-policy Rave (at the end of training simulation) = {}".format(Rave_stepseries[-1]))
    print("optimal theta_ON found = {}".format(theta_final[0]))
    print("optimal theta_OFF found = {}".format(theta_final[1]))
    print("Rave using the trained deterministic policy = {}".format(
        Rave_with_theta_final))
    
    if fixed_TransTime:
        envn_var_cont=copy.deepcopy(envn)
        envn_var_cont.fixed_TransTime=False
        Rave_fixtrain_varcontrol = envn_var_cont._RaveCalc(theta_final, 1, T_lowerbound = RL.Tlow,
                                          T_upperbound = RL.Tup, max_hour = trained_max_hour,
                                          deterministic = True, 
                                          PerTimeStep = RL.PerTimeStep,
                                          plots = False)
        
        print("Rave using deterministic policy (training: fixed transition time, deployed: variable transition) = {}".format(
        Rave_fixtrain_varcontrol))
    
    
    if HQ_fig: # plotting HQ and Latex-style figures
        
        mpl.rcParams['text.usetex'] = True
        fsize1=32
        fsize2=30

        fig, ax1 = plt.subplots(figsize=(16,6))
        color = 'tab:blue'
        ax1.set_xlabel(r'$\mathrm{time (hr)}$', fontsize=fsize1)
        ax1.set_ylabel(r'$\mathrm{Temperature}(^{\circ}C)$', fontsize=fsize1)
        ax1.plot(t_timeseries, T_timeseries, color=color)
        ax1.set_xlim(0, 160)
        for t in ax1.xaxis.get_major_ticks(): t.label.set_fontsize(fsize2)
        for t in ax1.yaxis.get_major_ticks(): t.label.set_fontsize(fsize2)
            
        axins = inset_axes(ax1, 2,3 , loc=1,bbox_to_anchor=(0.88, 0.75),bbox_transform=ax1.figure.transFigure) # no zoom
        axins.plot(t_timeseries, T_timeseries)
        x1, x2, y1, y2 = 110, 115, 11, 19 # specify the limits
        axins.set_xlim(x1, x2) # apply the x-limits
        axins.set_ylim(y1, y2) # apply the y-limits
        for t in axins.xaxis.get_major_ticks(): t.label.set_fontsize(fsize2)
        for t in axins.yaxis.get_major_ticks(): t.label.set_fontsize(fsize2)

        mark_inset(ax1, axins, loc1=2, loc2=3, fc="none", ec="red")
        plt.show()
        if save_fig: #saving plots
            fig.savefig('temp_time_COPDAC.png',dpi=200, bbox_inches='tight')
            fig.savefig('temp_time_COPDAC.pdf',bbox_inches='tight')
    
    
    output=[]
    output=[theta_final, t_stepseries, Rave_stepseries, Rave_with_theta_final]

    if fixed_TransTime:
        output.append(Rave_fixtrain_varcontrol)

    if return_arg:
        return output

env_gt = Simple1Room(Td=15.0)
RL = COPDAC_Q(Td=env_gt.Td,
              learning_rate_policy_mu=0.020,
              learning_rate_baselinefunc=0.050,
              learning_rate_actionvaluefunc=0.050,
              learning_rate_avereward=0.050,
              policy_update_freq=1,
              theta_init=[11, 19],
              T_upperbound=22,
              T_lowerbound=8,
              PerTimeStep=True)

COPDAC_Simple1Room(env_gt, RL, max_train_days=10, max_train_steps=3000,
                   fixed_TransTime=False, dt_TransStep=5*60,
                   training_plots=False, trained_max_hour=5*24.0,
                   trained_plots=False, HQ_fig=True, save_fig=True)

```

```python
# Here we show how the variable-time interval approach outperforms the fixed-time interval approach in terms of
# learning variance.

output_var=COPDAC_Simple1Room(env_gt, RL, max_train_days=10, max_train_steps=3000,
                   fixed_TransTime=False, dt_TransStep=5*60,
                   training_plots=False, trained_max_hour=5*24.0,
                   trained_plots=False, HQ_fig=False, save_fig=False,
                          return_arg=True)
output_fixed=COPDAC_Simple1Room(env_gt, RL, max_train_days=10, max_train_steps=3000,
                   fixed_TransTime=True, dt_TransStep=5*60,
                   training_plots=False, trained_max_hour=5*24.0,
                   trained_plots=False, HQ_fig=False, save_fig=False,
                          return_arg=True)


mpl.rcParams['text.usetex'] = True
AxesLabelFont = FontProperties(family='Arial', size=22)
TickLabelFont = FontProperties(family='Arial', size=22)

fig, ax = plt.subplots(figsize=(10,6))
ax.set_xlabel('time (hour)', fontproperties=AxesLabelFont)
ax.set_ylabel(r'Average reward rate $\bar{r}$ (unit/hour)', fontproperties=AxesLabelFont)
ax.plot(output_fixed[1], output_fixed[2], color='C3', label='constant time interval')
ax.plot(output_var[1], output_var[2], color='C0', linewidth=3.0, label='variable time interval')
plt.legend(fontsize=22)
a = plt.gca()
a.set_xticklabels(a.get_xticks(), fontproperties=TickLabelFont)
a.set_yticklabels(a.get_yticks(), fontproperties=TickLabelFont)
fig.tight_layout()

fig.savefig('ConstVar_time_COPDAC.png',dpi=200, bbox_inches='tight')
fig.savefig('ConstVar_time_COPDAC.pdf',bbox_inches='tight')
plt.show()
```

## D. Implementing COPDAC_Q_Ton and PPO on Simple1Room 
Here we implement a deterministic actor-critic policy defined by the *COPDAC_Q_Ton* class with variable transition intervals. We also implement the PPO from Stable Baselines3 implementation (we had to create our own gym-compatible environment class for the building room to implement the PPO on.)

```python
num_runs = 10
global policy_Ton_list, SPG_run
policy_Ton_list = []
SPG_run = [1]
class SimpleCallback(BaseCallback):
    """
    a simple callback that can only be called twice

    :param verbose: (int) Verbosity level 0: not output 1: info 2: debug
    """
    def __init__(self, verbose=0):
        super(SimpleCallback, self).__init__(verbose)
        global cum_rew, action, temp, hs, policy_Ton
        cum_rew = [0]
        action = []
        temp = []
        hs = []
        policy_Ton = []
        print('SPG run number:{i}'.format(i=len(SPG_run)))
    
    def _on_step(self):
        cum_rew.append(cum_rew[-1]+self.locals['rewards'][0])
        action.append(self.locals['actions'][0][0])
        temp.append(self.locals['obs_tensor'].numpy()[0][0])
        hs.append(self.locals['obs_tensor'].numpy()[0][1])
        act, st = self.model.predict(np.array([0.5, 1]), deterministic=True)
        policy_Ton.append(act)
    
    def _on_training_end(self):
        """
        This event is triggered before exiting the `learn()` method.
        """
        SPG_run.append(1)
        policy_Ton_list.append(policy_Ton)
        pass
    


env = Simple1Room_gym(Td=15.0,
                      relect=-1 * 0.12 * 10 / 3600,
                      rcom=-1 * 0.12 * 10 / 3600,
                      rsw=-1*1.1,
                      dt_TransStep=5,
                      time_max = 6)


eval_env = deepcopy(env)
total_timesteps = 100*24*60*60/env.dt_TransStep
policy_kwargs = dict(activation_fn=th.nn.ReLU, net_arch=[dict(pi=[1, 1], vf=[6, 6])])

for _ in range(num_runs):
    simple_callback = SimpleCallback()
    model = PPO("MlpPolicy", env,
                tensorboard_log="./PPO_SimpleRoom_gym/",
                learning_rate = 0.006,
                n_steps=72,
                batch_size=36,
                policy_kwargs=policy_kwargs,
                verbose=0)
    model.learn(total_timesteps=total_timesteps, callback=simple_callback)

policy_Ton_list_denorm = [[env.denormalize_action_func(env.Tonup,env.Tonlow,x) for x in a] for a in policy_Ton_list]
policy_Ton_runs = np.array(policy_Ton_list_denorm)
policy_Ton_mean = policy_Ton_runs.mean(0)
policy_Ton_std = policy_Ton_runs.std(0)
time_SPG = [i*env.dt_TransStep/3600 for i in range(int(total_timesteps))]

y1_SPG = policy_Ton_mean-policy_Ton_std
y2_SPG = policy_Ton_mean+policy_Ton_std    

    
env_gt = Simple1Room(Td=15.0,
                     relect=-1 * 0.12 * 10 / 3600,
                     rcom=-1 * 0.12 * 10 / 3600,
                     rsw=-1*1.1)
RL = COPDAC_Q_Ton(Td=env_gt.Td,
              learning_rate_policy_mu=0.020,
              learning_rate_baselinefunc=0.050,
              learning_rate_actionvaluefunc=0.050,
              learning_rate_avereward=0.050,
              policy_update_freq=1,
              theta_init=[10.5],
              T_upperbound=22,
              T_lowerbound=8,
              PerTimeStep=True)

max_train_days = 100
max_train_steps = 20000

policy_Ton_det_list = []
t_stepseries_list = []

for _ in range(num_runs):
    
    Rave_stepseries, theta_stepseries, w_stepseries, v_stepseries, t_stepseries,\
    t_timeseries,T_stepseries, T_timeseries, hs_stepseries, G_stepseries, G_timeseries\
    = train_deterministic(env_gt, RL, max_simulation_days = max_train_days,
                          max_steps = max_train_steps)
    
    theta_ON_stepseries = [item[0] for item in theta_stepseries]
    t_stepseries_hr = [x / 3600 for x in t_stepseries]
    
    policy_Ton_det_list.append(theta_ON_stepseries)
    t_stepseries_list.append(t_stepseries_hr)
    

num_runsteps_list = [len(x) for x in t_stepseries_list]
min_runsteps = min(num_runsteps_list)
index_minsteps = num_runsteps_list.index(min_runsteps)


policy_Ton_det_list_trunc = [x[:min_runsteps] for x in policy_Ton_det_list]
policy_Ton_det_arr = np.array(policy_Ton_det_list_trunc)


policy_Ton_det_mean = policy_Ton_det_arr.mean(0)
policy_Ton_det_std = policy_Ton_det_arr.std(0)
time = t_stepseries_list[index_minsteps][:min_runsteps]

y1 = policy_Ton_det_mean-policy_Ton_det_std
y2 = policy_Ton_det_mean+policy_Ton_det_std

Ton_optimal = len(time_SPG)*[12.4]

mpl.rcParams['text.usetex'] = True
AxesLabelFont = FontProperties(family='Arial', size=22)
TickLabelFont = FontProperties(family='Arial', size=22)

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(time_SPG, Ton_optimal, 'k--', linewidth=2, label='optimal temperature')
ax1.plot(time, policy_Ton_det_mean, 'k-', color='#316a9a', label='proposed DPG')
ax1.fill_between(time, y1.ravel(), y2.ravel(), alpha=0.4, facecolor='#649dcd')
ax1.plot(time_SPG, policy_Ton_mean, 'k-', color='#CC4F1B',label='PPO (SPG)')
ax1.fill_between(time_SPG, y1_SPG.ravel(), y2_SPG.ravel(), alpha=0.3, facecolor='#FF9848')

ax1.set_xlabel('time (hr)', fontproperties=AxesLabelFont)
ax1.set_ylabel('switching temperature: $T_{\mathrm{ON}}(^{\circ}C)$', fontproperties=AxesLabelFont)



plt.legend(fontsize=22, loc = 'lower right')
a = plt.gca()
a.set_xticklabels(a.get_xticks(), fontproperties=TickLabelFont)
a.set_yticklabels(a.get_yticks(), fontproperties=TickLabelFont)
fig.tight_layout()
plt.show()

fig.savefig('DPG_SPG.png',dpi=400, bbox_inches='tight')
fig.savefig('DPG_SPG.pdf',bbox_inches='tight')
```
